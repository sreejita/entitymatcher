```python
import py_entitymatching as em
```

```python
A = em.read_csv_metadata('amazon_books_utf8.csv')
B = em.read_csv_metadata('walmart_books_utf8.csv')
```

```python
A['ID'] = range(0, len(A))
em.set_key(A, 'ID')
```

```python
B['ID'] = range(0, len(B))
em.set_key(B, 'ID')
```

```python
ob = em.OverlapBlocker()
C = ob.block_tables(A, B, 'Author', 'Author',word_level=True, overlap_size=2, l_output_attrs=['Name','Author','ISBN10'], r_output_attrs=['Name','Author','ISBN10'] )
```

```python
D = ob.block_candset(C, 'Name', 'Name',overlap_size=2,word_level=True, rem_stop_words=True)
```

```python
D
```

```python
E = ob.block_candset(D, 'ISBN10', 'ISBN10',overlap_size=1,word_level=True, rem_stop_words=True)
```

```python
E
```

```python
 debug_name1 = em.debug_blocker(D, A, B)
```

```python
S = em.sample_table(D, 30)
```

```python
debug_name1[['ltable_Name','ltable_ISBN10','ltable_Author','rtable_Name','rtable_ISBN10','rtable_Author']]
```

```python
labelled = em.label_table(S, label_column_name='gold_labels')
```

```python
labelled
```

```python
A['Name'] = A['Name'].astype('str') 
B['Name'] = B['Name'].astype('str')
```

```python
match_f = em.get_features_for_matching(A,B)
```

```python
match_f
```

```python
train_test = em.split_train_test(labelled, train_proportion=0.5)
```

```python
devel_set = train_test['train']
eval_set = train_test['test']
```

```python
H = em.extract_feature_vecs(labelled, feature_table=match_f,attrs_after=['gold_labels'])
```

```python
imputed = em.impute_table(H, exclude_attrs=['_id', 'ltable_ID', 'rtable_ID'], strategy='mean')
```

```python
train_test = em.split_train_test(imputed, train_proportion=0.5)
```

```python
devel_set = train_test['train']
eval_set = train_test['test']
```

```python
dt = em.DTMatcher(max_depth=5)
```

```python
dt.fit(table=devel_set, exclude_attrs=['_id', 'ltable_ID', 'rtable_ID'], target_attr='gold_labels')
```

```python
dt.predict(table=eval_set, exclude_attrs=['_id', 'ltable_ID', 'rtable_ID'], target_attr='predicted_labels', return_probs=True, probs_attr='proba', append=True,
inplace=True)
```
